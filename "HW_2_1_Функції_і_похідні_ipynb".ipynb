1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
x = sp.Symbol('x')
f = 10*x - 5
f_derivative = sp.diff(f, x)

print("Похідна функції f(x) = 10x - 5:")
print(f_derivative)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Діапазон значень x
x = np.linspace(0, 10, 100)

# Обчислення значень функції та її похідної
f_values = 10*x - 5
f_derivative_values = 10 * np.ones_like(x)

# Побудова графіків
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, f_values)
plt.title("Графік функції f(x) = 10x - 5")
plt.xlabel("Кількість годин")
plt.ylabel("Оцінка")

plt.subplot(1, 2, 2)
plt.plot(x, f_derivative_values)
plt.title("Графік похідної f'(x) = 10")
plt.xlabel("Кількість годин")
plt.ylabel("Похідна")

plt.tight_layout()
plt.show()
```

Аналізуючи графіки, можна побачити, що оцінка лінійно зростає зі збільшенням кількості годин навчання, а похідна функції постійна і дорівнює 10, що означає, що оцінка збільшується на 10 одиниць за кожну додаткову годину навчання.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x = sp.Symbol('x')
f1 = sp.sin(x)
f1_derivative = sp.diff(f1, x)
f2 = sp.log(x)
f2_derivative = sp.diff(f2, x)

print("Похідна функції f(x) = sin(x):")
print(f1_derivative)
print("Похідна функції f(x) = ln(x):")
print(f2_derivative)
```

```python
# Діапазон значень x
x = np.linspace(0.1, 10, 100)

# Обчислення значень функцій та їхніх похідних
f1_values = np.sin(x)
f1_derivative_values = np.cos(x)

f2_values = np.log(x)
f2_derivative_values = 1/x

# Побудова графіків
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x, f1_values)
plt.title("Графік функції f(x) = sin(x)")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.subplot(2, 2, 2)
plt.plot(x, f1_derivative_values)
plt.title("Графік похідної f'(x) = cos(x)")
plt.xlabel("x")
plt.ylabel("f'(x)")

plt.subplot(2, 2, 3)
plt.plot(x, f2_values)
plt.title("Графік функції f(x) = ln(x)")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.subplot(2, 2, 4)
plt.plot(x, f2_derivative_values)
plt.title("Графік похідної f'(x) = 1/x")
plt.xlabel("x")
plt.ylabel("f'(x)")

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.Symbol('x')
f = x**2 - 5*x + 1
f_derivative = sp.diff(f, x)
print("Функція f(x):", f)
print("Похідна f'(x):", f_derivative)
```

```python
x_values = np.linspace(0, 100, 101)
f_values = [f.subs(x, x_val) for x_val in x_values]
f_derivative_values = [f_derivative.subs(x, x_val) for x_val in x_values]

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_values, f_values)
plt.title("Функція f(x) = x^2 - 5x + 1")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Загальні витрати (f(x))")

plt.subplot(1, 2, 2)
plt.plot(x_values, f_derivative_values)
plt.title("Похідна f'(x)")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Похідна (f'(x))")

plt.tight_layout()
plt.show()
```
